```python
import pymongo
import pandas as pd
```

```python
# connect to the MongoDB
connection = pymongo.MongoClient("mongodb+srv://")

db = connection.rsmp     # attach to db
collection = db.docs      # specify the collection
```

### Юридические лица

```python
pipeline =[
    {'$match':{'ИПВклМСП':{'$exists':0}}},
    {'$project': {'_id':0,
                 'ОргВклМСПИННЮЛ':'$ОргВклМСП.ИННЮЛ',
                 'ОргВклМСПНаимОргСокр':'$ОргВклМСП.НаимОргСокр',
                 'ОргВклМСПНаимОрг':'$ОргВклМСП.НаимОрг',
                 'КодРегион':'$СведМН.КодРегион',
                 'РайонТип':'$Район.Тип',
                 'РайонНаим':'$Район.Наим',
                 'НаселПунктТип':'$НаселПункт.Тип',
                 'НаселПунктНаим':'$НаселПункт.Наим',
                 'ГородТип':'$Город.Тип',
                 'ГородНаим':'$Город.Наим',
                 'СвОКВЭДОснКодОКВЭД':'$СвОКВЭДОсн.КодОКВЭД',
                 'СвОКВЭДОснНаимОКВЭД':'$СвОКВЭДОсн.НаимОКВЭД',
                 'СвОКВЭДОснВерсОКВЭД':'$СвОКВЭДОсн.ВерсОКВЭД',
                 'ИдФайл':1,
                 'ИдДок':1}
    }
]
```

```python
cursor = collection.aggregate(pipeline)
# Expand the cursor and construct the DataFrame
df =  pd.DataFrame(list(cursor))
df = df[list(pipeline[1]['$project'].keys())[1:]] # reoder columns as in pipeline
df.to_excel('ul_region_14.xls', index=False)
```
